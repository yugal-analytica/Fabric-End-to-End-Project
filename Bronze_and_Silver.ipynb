#### Deduplication 

```python
window= Window.partitionBy("SaleID").orderBy("SaleDate")
df_june_active= df_june_active.withColumn("rn", row_number().over(window))\
                                .filter(col("rn")==1)\
                                .drop("rn")
```

```python
df_july_active= df_july_active.withColumn("rn", row_number().over(window))\
                                .filter(col("rn")==1)\
                                .drop("rn")
```

```python
df_june_active.write.format("delta").mode("overwrite").saveAsTable("June_Sales")
```

```python
df_july_active.write.format("delta").mode("overwrite").saveAsTable("July_Sales")
```
